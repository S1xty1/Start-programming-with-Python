```python
#导入相关库
import copy
import pandas as pd
import numpy as np
#读取数据
data=pd.read_excel('D:\sen\lianghua.xls')
print(data)
 
label_need=data.keys()[1:]
print(label_need)
data1=data[label_need].values
print(data1)
[m,n]=data1.shape

#计算信息熵
p=copy.deepcopy(data1)
for j in range(0,n):
    p[:,j]=data1[:,j]/sum(data1[:,j])
print(p)
E=copy.deepcopy(data1[0,:])
for j in range(0,n):
    temp=np.nan_to_num(p[:,j]*np.log(p[:,j]))
    E[j]=-1/np.log(m)*sum(temp)


print(E)

 # 计算权重
w=(1-E)/sum(1-E)
f1 =open("D:\sen\熵权.xls", 'w',encoding='utf-8',errors='ignore')
print(w,file=f1)
f2 =open("D:\sen\分数.xls", 'w',encoding='utf-8',errors='ignore')
#计算得分
s=np.dot(data1,w)
Score=100*s/max(s)
for i in range(0,len(Score)):
    print(f"{Score[i]}",file=f2)
```
