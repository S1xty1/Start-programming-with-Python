```python
import csv
#定义列表names和number_1
data_1=[]
number_1 = 0

with open('e:/data_with_noise.csv', encoding='utf-8-sig') as line_1:
    #依次读取CSV文件的每一行
    for line_2 in line_1.readlines():
        #strip() 方法用于移除字符串头尾指定的字符（默认为空格或换行符）或字符序列
        line_2 = line_2.strip()
        #split() 通过指定分隔符对字符串进行切片，这里指定','，而“-1”表示分隔所有
        list_1 = line_2.split(',', -1)
        #重置number_1所表示的数字
        number_1 = 0
        #依次读取一行中的每个元素，并用append()添加到列表data_1中
        while number_1 <len(list_1):
            data_1.append(list_1[number_1])
            number_1=number_1+1
    print(data_1)
```

```python
data_1=data_1[3:]
print(data_1)
```

```python
x=[]
y=[]
for i in range(10000):
    x.append(float(data_1[1+3*i]))
for i in range(10000):
    y.append(float(data_1[2+3*i]))
```

```python
  
import matplotlib.pyplot as plt
plt.scatter(x,y)
```
