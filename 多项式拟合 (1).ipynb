```python
import csv
#namesnumber_1
data_1=[]
number_1 = 0

with open('e:/data_with_noise.csv', encoding='utf-8-sig') as line_1:
    #CSV
    for line_2 in line_1.readlines():
        #strip() 
        line_2 = line_2.strip()
        #split() ','-1
        list_1 = line_2.split(',', -1)
        #number_1
        number_1 = 0
        #append()data_1
        while number_1 <len(list_1):
            data_1.append(list_1[number_1])
            number_1=number_1+1
    print(data_1)
```

```python
data_1=data_1[3:]
print(data_1)
```

```python
x=[]
y=[]
for i in range(10000):
    x.append(float(data_1[1+3*i]))
for i in range(10000):
    y.append(float(data_1[2+3*i]))
```

```python
  
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures as pf
import numpy as np

xmat=np.array(x).reshape(-1,1)
ymat=np.array(y).reshape(-1,1)


poly_features = pf(degree=9,include_bias=False)
x_poly=poly_features.fit_transform(xmat)
x_poly[0]

from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x_poly,y)
lin_reg.intercept_,lin_reg.coef_

X_new=np.linspace(-50,50,10000).reshape(10000,1)
X_new_poly=poly_features.transform(X_new)
y_new=lin_reg.predict(X_new_poly)
plt.scatter(x,y)
plt.plot(X_new,y_new,'r-')
plt.show
#7-9>=10<=6
```
