```python
count=[0]*6
import random
portions=[[],[],[],[],[],[]]
for j in range(0,1000):
    for i in range(0,50):
        count[random.randint(1,6)-1]+=1
    for i in range(0,6):
        portions[i].append(count[i]/sum(count))
print(count)
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots()

x = list(range(1000))
plt.plot(x,portions[0],'r',x,portions[1],'b',x,portions[2],'g',x,portions[3],'',x,portions[4],'',x,portions[5],'',)
plt.show()
```

```python
import random
count = 0
last_3=[0,0,0]
for i in range(0,10000):
    value=random.randint(1,6)
    last_3.pop(0)
    last_3.append(value)
    if last_3==[6,6,6]:
        count+=1
print(count)
print(count/10000)
```

```python
import random
count = [0]*6
last_3=[0,0,0]
for i in range(0,1000000):
    value=random.randint(1,6)
    last_3.pop(0)
    last_3.append(value)
    if last_3[0]==last_3[1]==last_3[2]:
        count[value-1]+=1
print(count)
```

```python
str = 'Hello,this is a test used to count appearence of vowel letters in a sentence.'
count = dict.fromkeys('aeiou',0) #给dictionary一个列表（aeiou），作为它的key，查找的内容为其累计个数，目前初始化为0
for ch in str:
    if ch in 'aeiou':
        count[ch]+=1
print(count)
```
