水滴实验：定量分析
每次5ml水滴
温度（度）：150, 160, 170, 180, 190, 200, 210, 220, 230, 240, 250, 260, 270, 280, 290, 300
完全消失时间（秒）：0, 0, 2.0, 2.0, 2.7, 3.0, 4.0, 3.4, 4.0, 4.0, 5.2, 5.0, 10.0, 8.0, 11.0, 5.0

```python
x1 = [170, 180, 190, 200, 210]
y1 = [2.0, 2.0, 2.7, 3.0, 4.0]
x2 = [220, 230, 240, 250, 260, 270, 280, 290]
y2 = [3.4, 4.0, 4.0, 5.2, 5.0, 10.0, 8.0, 11.0]

plt.plot(x1+x2,y1+y2,'.')

plt.savefig('0.jpg',dpi=200, bbox_inches='tight') # 解决图片不清晰，不完整的问题

plt.show()


```

曲线拟合


```python
z1 = np.polyfit(x1, y1, 2)
z2 = np.polyfit(x2, y2, 2) #用3次多项式拟合，输出系数从高到0

print(z1, z2)

p1 = np.poly1d(z1) #使用次数合成多项式
p2 = np.poly1d(z2)

y_fit1 = p1(x1+[220]) #拟合曲线上y的值
y_fit2 = p2([200, 210]+x2)

plt.plot(x1+x2,y1+y2,'.')

plt.plot(x1+[220],y_fit1)
plt.plot([200, 210]+x2,y_fit2)


plt.savefig('1.jpg',dpi=200, bbox_inches='tight') # 解决图片不清晰，不完整的问题

plt.show()
```
